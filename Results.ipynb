```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data= pd.read_csv("../RR1.csv")
data['SMA30'] = data['score'].rolling(100).mean()
print(data['score'][99420:99440])
f = plt.figure()
f.set_figwidth(40)
f.set_figheight(20)
plt.plot(data['SMA30'])
plt.grid()
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.yticks(np.arange(-10, 11, 2))
plt.show()
```

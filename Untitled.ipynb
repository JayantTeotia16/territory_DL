```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data= pd.read_csv("results_territory/rewards_files/5e-05_Adam_1000000_1_100_1000_0_4_256_DQN_PER_0.5_False_3_10.csv")
data['SMA30'] = data['score'].rolling(1000).mean()
data1 = pd.read_csv("results_territory/rewards_files/game1.csv")
data['A30'] = data1['score'].rolling(1000).mean()
#print(data['SMA30'][344500:])
f = plt.figure()
f.set_figwidth(40)
f.set_figheight(20)
plt.plot(data['SMA30'], color='r', label='RR1')
plt.plot(data['A30'], color='b', label='RR')
plt.grid()
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.yticks(np.arange(-10, 11, 1))
plt.show()

```
